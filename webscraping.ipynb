```python
import requests 
from bs4 import BeautifulSoup
page = requests.get("https://en.wikipedia.org/wiki/List_of_cities_and_largest_towns_in_Myanmar")
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
print(soup.find('table'))
```

```python
table = soup.find('table')
trs = table.find_all('tr')

for tr in trs:
  for td in tr.find_all('td'):
    print(td)
```

```python
table = soup.find('table')
trs = table.find_all('tr')
rows = []
for tr in trs:
  for td in tr.find_all('td'):
    rows.append([td.get_text(strip=True) for td in tr.find_all('td')])

rows
```

```python
table = soup.find('table')
trs = table.find_all('tr')
rows = []
for tr in trs:
  for td in tr.find_all('td'):
    rows.append([td.get_text(strip=True) for td in tr.find_all('td')])

rows
```

```python
import pandas as pd
dftable = pd.DataFrame(rows[:], columns=rows[0])
dftable
```

```python
a = dftable.to_csv('myanmar_city_list.csv')
a
```
